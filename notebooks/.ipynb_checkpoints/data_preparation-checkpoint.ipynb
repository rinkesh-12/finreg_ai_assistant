```python
import os
from dotenv import load_dotenv

load_dotenv()  # take environment variables from .env.
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
## Data Ingestion
from langchain_community.document_loaders import PyPDFLoader
import re
```

```python
# Load the PDF file
try:
    loader = PyPDFLoader("../data/rbi.pdf")
    docs = loader.load()
    print(f"Loaded {len(docs)} pages from PDF")
except Exception as e:
    print(f"Error loading PDF file: {e}")
    docs = []
```

```python
from langchain.schema import Document

# Clean each page's text
cleaned_docs = []
for d in docs:
    # Normalize whitespace
    text = d.page_content.replace("\n", " ")
    text = " ".join(text.split())

    if text.strip():  # only keep non-empty text
        cleaned_docs.append(Document(page_content=text, metadata=d.metadata))
# print(f"Number of cleaned documents: {len(cleaned_docs)}")
print(f"Cleaned documents: {len(cleaned_docs)} pages")
```

```python
# hierarchical splitter
from langchain.text_splitter import RecursiveCharacterTextSplitter # CharacterTextSplitter

# Get total length of all docs
total_length = sum(len(doc.page_content) for doc in docs)

# Set chunk_size to total_length if it's less than your default, else use default
default_chunk_size = 2500
chunk_size = total_length if total_length < default_chunk_size else default_chunk_size

splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000, # chunk_size, 
    chunk_overlap=200, 
    separators=["\n\n", "\n", ".", " ", ""]
)

chunks = splitter.split_documents(docs)

print(f"Total pages: {len(docs)}")
print(f"Chunks created: {len(chunks)}\n")
```

```python
for i, c in enumerate(chunks[:5]):
    print(f"\n--- Chunk {i+1} ---\n{c.page_content[:400]}...")
```

```python
# Print the first chunk to verify content
print(chunks[0])  
print(chunks[0].page_content)
```

```python
# Vector Embedding and Vector Store
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# Initialize the embedding model (local, free)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Create the vector store from your chunks
vectordb = FAISS.from_documents(chunks, embeddings)
vectordb.save_local("../data/faiss_index")

# Check the vector store
print(vectordb)
```

```python
query = "What does conducting financial activity as “principal business” mean?"
result = vectordb.similarity_search(query)
print(result[0])
```

```python
# Design ChatPromp Template
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

system = SystemMessagePromptTemplate.from_template(
    "You are an expert financial assistant. Answer the following question based ONLY on the provided context. "
    "If the answer cannot be found in the context, state that you cannot find the answer. "
    "Think step by step before providing a detailed answer."
)

human = HumanMessagePromptTemplate.from_template(
    "<context>\n{context}\n</context>\nQuestion: {input}"
)

prompt = ChatPromptTemplate.from_messages([system, human])
```

```python
## Retirever And Chain With LangChain and Gemini Pro

from langchain_community.llms import Ollama
from langchain_google_genai import ChatGoogleGenerativeAI
import os

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY") # "AIzaSyAlMSO9hWfZ1k6CbIQ2AS8ooqq5-2oKpuo"

# LLM (Gemini free tier)
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
llm
```

```python
## Chain
from langchain.chains.combine_documents import create_stuff_documents_chain
chain = create_stuff_documents_chain(llm, prompt=prompt)
chain
```

```python
"""
Retrievers: A retriever is an interface that returns documents given
 an unstructured query. It is more general than a vector store.
 A retriever does not need to be able to store documents, only to 
 return (or retrieve) them. Vector stores can be used as the backbone
 of a retriever, but there are other types of retrievers as well. 
 https://python.langchain.com/docs/modules/data_connection/retrievers/
"""

retriever = vectordb.as_retriever()
retriever
```

```python
"""
Retrieval chain:This chain takes in a user inquiry, which is then
passed to the retriever to fetch relevant documents. Those documents 
(and original inputs) are then passed to an LLM to generate a response
https://python.langchain.com/docs/modules/chains/
"""

from langchain.chains import create_retrieval_chain
retrieval_chain = create_retrieval_chain(retriever, chain)
retrieval_chain
```

```python
result = retrieval_chain.invoke({"input": "Investment through Alternative Investment Funds - Calculation of NOF of an NBFC"})
print(result["answer"])
```
